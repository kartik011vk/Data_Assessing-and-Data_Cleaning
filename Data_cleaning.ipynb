# Making copy of all the datasets

```python
adverse_reactions = adverse_reactions.copy()
treatment_copy = treatment.copy()
treatment_cut_copy = treatment_cut.copy()
patients_copy = patients.copy()
```

```python
patients_copy[patients_copy['address'].isnull()]
patients_copy.fillna("unknown", inplace = True)
```

```python
patients_copy[patients_copy['address'].isnull()]
```

```python
treatment.head()
treatment_copy["hba1c_change"] = treatment_copy["hba1c_start"] - treatment_copy["hba1c_end"]
treatment_copy.head()
```

```python
treatment_cut_copy["hba1c_change"] = treatment_cut_copy["hba1c_start"] - treatment_cut_copy["hba1c_end"]
treatment_copy.head()
```

```python
patients_copy.head()
```

```python
patients_copy["Email"] = patients_copy["contact"].str.extract(
    r'([A-Za-z]+[A-Za-z0-9._%+-]*@[A-Za-z0-9.-]+\.[A-Za-z]{2,})'
)

patients_copy["Phone"] = patients_copy["contact"].str.extract(
    r'(\+?\d[\d\s\-\(\)]{7,}\d)'
)
```

```python
patients_copy.head()
```

```python
patients_copy = patients_copy.drop("contact", axis = 1)
```

```python
patients_copy.head()
```

```python
treatments_df = pd.concat([treatment_copy, treatment_cut_copy], ignore_index = True)
```

```python
treatments_df.head()
```

```python
treatments_df = treatments_df.melt(id_vars=['given_name', 'surname' ,'hba1c_start', 'hba1c_end','hba1c_change'],var_name='type',value_name='dosage_range')
```

```python
treatments_df
```

```python
treatments_df = treatments_df[treatments_df['dosage_range'] != '-']
```

```python
treatments_df['dosage_start'] = treatments_df['dosage_range'].str.split('-').str.get(0)
treatments_df['dosage_end'] = treatments_df['dosage_range'].str.split('-').str.get(1)
```

```python
treatments_df.head()
```

```python
treatments_df
```

```python
treatments_df = treatments_df.merge(adverse_reactions , how ='left', on=['given_name','surname'])
treatments_df
```

```python
treatments_df["adverse_reaction"].fillna("No reaction", axis = 0, inplace = True)
```
